```python
import folium
from folium.features import GeoJsonPopup

# Convert the merged DataFrame to a GeoDataFrame
gdf_merged = gpd.GeoDataFrame(merged_df, geometry='geometry')

# Create a folium map centered on Breda
m = folium.Map(location=[51.5831, 4.7762], zoom_start=12)

# Convert GeoDataFrame to GeoJSON
gdf_json = gdf_merged.to_json()

# Create a function to generate the popup content for a neighborhood
def generate_popup_content(properties):
    content = "<h4>Nuisance Report</h4>"
    content += "<table>"
    content += "<tr><th>Type</th><th>Number</th></tr>"
    for column, value in properties.items():
        if column not in ['geometry', 'neighborhood']:
            content += f"<tr><td>{column}</td><td>{value}</td></tr>"
    content += "</table>"
    return content

# Add GeoJSON data to the map
folium.GeoJson(
    gdf_json,
    name='Nuisance Data',
    popup=folium.GeoJsonPopup(
        fields=['Total nuisance registrations', 'Nuisance by confused person', 'Youth nuisance report', 'Nuisance due to alcohol/drugs', 'Nuisance drifters', 'Public intoxication', 'Noise nuisance catering', 'Noise nuisance event', 'Other noise nuisance'],
        aliases=['Total', 'Confused', 'Youth', 'Alcohol/Drugs', 'Drifters', 'Intoxication', 'Catering', 'Event', 'Other'],
        localize=True,
        labels=True,
        style="background-color: yellow; color: black; font-weight: bold;",
        parse_html=False
    )
).add_to(m)

# Display the map
m
```
