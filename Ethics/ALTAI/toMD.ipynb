```python
import mammoth
import markdown

# Load the Word document
with open("Accountability.docx", "rb") as docx_file:
    result = mammoth.convert_to_html(docx_file)
    html = result.value

# Convert HTML to Markdown
markdown_text = markdown.markdown(html)

# Turn headings into ##
markdown_text = markdown_text.replace("# ", "## ")
markdown_text = markdown_text.replace("## ", "### ")
markdown_text = markdown_text.replace("### ", "#### ")
markdown_text = markdown_text.replace("#### ", "##### ")
markdown_text = markdown_text.replace("##### ", "###### ")

# Make bold text bold
markdown_text = markdown_text.replace("<strong>", "**")
markdown_text = markdown_text.replace("</strong>", "**")

# Save the result
with open("example.md", "w") as md_file:
    md_file.write(markdown_text)
```

```python
import mammoth
import markdown
import re

with open("Accountability.docx", "rb") as docx_file:
    result = mammoth.convert_to_html(docx_file)
    html = result.value

html = html.encode("ascii", "ignore").decode()

html = re.sub(r'[^\x00-\x7F]+', '', html)

markdown_text = markdown.markdown(html)

markdown_text = markdown_text.replace("# ", "## ")
markdown_text = markdown_text.replace("## ", "### ")
markdown_text = markdown_text.replace("### ", "#### ")
markdown_text = markdown_text.replace("#### ", "##### ")
markdown_text = markdown_text.replace("##### ", "###### ")

markdown_text = markdown_text.replace("<ul>", "")
markdown_text = markdown_text.replace("</ul>", "")
markdown_text = markdown_text.replace("<li>", "")
markdown_text = markdown_text.replace("</li>", "")

markdown_text = re.sub(r"\*\* +", "**", markdown_text)
markdown_text = re.sub(r"\*\*(.*?)\*\*", r"\n\n**\1**\n\n", markdown_text)

with open("example.md", "w") as md_file:
    md_file.write(markdown_text)
```
