```python
df_sort = df.sort_values('proximity_score', ascending=False)
```

```python
col1 = 'proximity_score'
col2 = 'distance_from_centre_km'

fig, ax1 = plt.subplots()

ax1.set_xlabel('neighborhood')
ax1.set_ylabel(col1, color=color1)
ax1.plot(df_sort['neighborhood'], df_sort[col1])
ax1.tick_params(axis='y', color=color1)
ax1.set_title(f'{col1} and {col2}')
plt.tick_params(axis='x',
                which='both',
                labelbottom=False)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

ax2.set_ylabel(col2, color=color2)  # we already handled the x-label with ax1
ax2.plot(df_sort['neighborhood'], df_sort[col2], color=color2)
ax2.tick_params(axis='y', labelcolor=color2)

# fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()

df[[col1, col2]].corr(method='spearman')
```
